## Análisis gráfico

###Edad

````r
g=ggplot(Mall_Customers,aes(x= Age))
g=g+geom_histogram(aes(y = ..density..),bins = 20,color = 'black', fill = "gray")
g=g+geom_density(fill="green", alpha = .2)
g=g+ylab("fX(x)")
g=g+xlab("Edad")
g=g+theme_minimal()
g
````

###Ingreso

````r
g=ggplot(Mall_Customers,aes(x= Annual.Income..k..))
g=g+geom_histogram(aes(y = ..density..),bins = 20,color = 'turquoise4', fill = "gray")
g=g+geom_density(fill="green", alpha = .2)
g=g+ylab("fX(x)")
g=g+xlab("Ingreso")
g=g+theme_minimal()
g
````

*italicized text*#K-MEAN

##Determinar la cantidad de clusters

##Estimar Cluster

# Nueva sección

````r
km.out=kmeans(Mall_Customers[,3:5],6,nstart = 50)
km.out
````

##Graficar

[link text](https://)###Ingreso vs Spending Score

````r
g=ggplot(Mall_Customers, aes(x = Annual.Income..k.., y = Spending.Score..1.100.))
g=g+geom_point(stat = "identity", aes(color = as.factor(km.out$cluster)))
g=g+scale_color_discrete(name=" ",breaks=c("1", "2", "3", "4", "5","6"), labels=c("Cluster 1", "Cluster 2", "Cluster 3", "Cluster 4", "Cluster 5","Cluster 6"))
g=g+ggtitle("Segmentos de mercado", subtitle = "K-means Clustering")
g=g+ylab("Spending Score")
g=g+xlab("Ingreso")
g=g+theme_minimal()
g




````

###Edad vs Spending Score

````r
g=ggplot(Mall_Customers, aes(x =Age, y = Spending.Score..1.100.))
g=g+geom_point(stat = "identity", aes(color = as.factor(km.out$cluster)))
g=g+scale_color_discrete(name=" ",breaks=c("1", "2", "3", "4", "5","6"), labels=c("Cluster 1", "Cluster 2", "Cluster 3", "Cluster 4", "Cluster 5","Cluster 6"))
g=g+ggtitle("Segmentos de mercado", subtitle = "K-means Clustering")
g=g+ylab("Spending Score")
g=g+xlab("Edad")
g=g+theme_minimal()
g
````

Edad vs Ingreso Anual

````r
g=ggplot(Mall_Customers, aes(x =Age, y = Annual.Income..k..))
g=g+geom_point(stat = "identity", aes(color = as.factor(km.out$cluster)))
g=g+scale_color_discrete(name=" ",breaks=c("1", "2", "3", "4", "5","6"), labels=c("Cluster 1", "Cluster 2", "Cluster 3", "Cluster 4", "Cluster 5","Cluster 6"))
g=g+ggtitle("Segmentos de mercado", subtitle = "K-means Clustering")
g=g+ylab("Ingreso")
g=g+xlab("Edad")
g=g+theme_minimal()
g
````


##Mapenado los segmentos de mercado en los datos originales

````r
o=order(km.out$cluster)
datos = cbind(Mall_Customers,km.out$cluster[o] )

````
